```python
mask = (data.Temp >= 00) & (data.Temp <= 45)
workingRange = data.loc[mask]

"reference @https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html "

y= workingRange.Temp
x= workingRange.Voltage
z = np.polyfit(x, y, 6)
p = np.poly1d(z)
print(p)
print(p(2.16))
```

```python
plt.plot(workingRange.Voltage, p(workingRange.Voltage), '-', workingRange.Voltage, workingRange.Temp, '--')
plt.xlim(1, 3.5)
plt.show()
```
